```python
import xgboost as xgb
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("../datasets/churn.csv")
```

```python
dataset.head()
```

```python
churn_data = dataset.iloc[:, 4:]
churn_data.head()
```

```python
churn_data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for cols in churn_data.columns:
    if churn_data[cols].dtypes == "object":
        churn_data[cols] = le.fit_transform(churn_data[cols])
    elif churn_data[cols].dtypes == "bool":
        churn_data[cols] = le.fit_transform(churn_data[cols])
        
churn_data.head()
```

```python
X, y = churn_data.iloc[:,:-1], churn_data.iloc[:,-1]
X_train,X_test,y_train,y_test= train_test_split(X, y, test_size=0.2, random_state=123)
xg_cl = xgb.XGBClassifier(objective='binary:logistic', n_estimators=10, seed=123)
xg_cl.fit(X_train,y_train)
preds = xg_cl.predict(X_test)
accuracy = float(np.sum(preds==y_test))/y_test.shape[0]
print("accuracy: %f" % (accuracy))
```
